```python
#'질문내용_html' 컬럼을 보니 어떤 컬럼은 <html> 태그부터 시작하고, 어떤 컬럼은 <body> 컬럼 안쪽의 내용만 기록되어 있음
# 그래서 <body> 안의 내용만 정리하는 함수를 만들어서 '질문내용_body' 컬럼에 저장해본다.
```

```python
# 새로운 '내용_body' 컬럼 생성
def extract_body(content):
    if str(content).startswith('<html'):
        soup = BeautifulSoup(content, 'html.parser')
        body = soup.body
        return str(body) if body else ''
    else:
        return content

# '내용_body' 컬럼에 적용
df_q1['내용_body'] = df_q1['내용2'].apply(extract_body)

# 결과 확인
print(df_q1['내용_body'])
```

```python
# 새로운 '내용_body2' 컬럼 생성
def ensure_body_tags(content):
    content = str(content)
    if not content.startswith('<body>'):
        return f"<body>{content}</body>"
    return content

# '내용_body2' 컬럼에 적용
df_q1['내용_body2'] = df_q1['내용_body'].apply(ensure_body_tags)

# 결과 확인
print(df_q1['내용_body2'])
```

```python
# 테이블 추출 함수 정의
def extract_table(content):
    soup = BeautifulSoup(content, 'html.parser')
    table = soup.find('table')  # 첫 번째 <table> 태그 추출
    return str(table) if table else ''  # <table> 태그가 없으면 빈 문자열 반환

# '내용_body2_table' 컬럼 생성
df_q1['내용_body2_table'] = df_q1['내용_body2'].apply(extract_table)

# 결과 확인
print(df_q1['내용_body2_table'])
```

```python
def table_to_text(content):
    soup = BeautifulSoup(content, 'html.parser')
    table = soup.find('table')
    
    if not table:
        return ''  # 테이블이 없으면 빈 문자열 반환
    
    rows = table.find_all('tr')  # 모든 행 추출
    text_rows = []
    
    for row in rows:
        cells = row.find_all(['td', 'th'])  # 각 행의 셀(td 또는 th) 추출
        text_row = '|'.join(cell.get_text(strip=True) for cell in cells)  # 셀을 '|'로 구분
        text_rows.append(text_row)
    
    return '\n'.join(text_rows)  # 행을 줄바꿈으로 구분

# '내용_body2_table2' 컬럼 생성
df_q1['내용_body2_table'] = df_q1['내용_body2'].apply(table_to_text)

# 결과 확인
print(df_q1['내용_body2_table'])
```

```python
def extract_body_as_text_with_table(html):
    soup = BeautifulSoup(html, 'html.parser')
    
    # 테이블 부분을 먼저 처리
    def extract_table_text(table):
        rows = []
        for tr in table.find_all('tr'):
            cells = [td.get_text(strip=True) for td in tr.find_all(['td', 'th'])]
            rows.append("|".join(cells))  # 열을 탭으로 구분
        return "\n".join(rows)  # 행을 줄바꿈으로 구분

    final_text = []

    # <body> 태그에서 모든 자식 요소를 순차적으로 처리
    for element in soup.body.children:
        if element.name == 'table':
            # <table>일 경우, 테이블 텍스트로 변환
            final_text.append(extract_table_text(element))
        elif element.name in ['p', 'br', 'ul', 'ol', 'li']:
            # 텍스트가 포함된 태그(p, br, ul, ol, li 등)에 대해서는 그대로 텍스트 추가
            final_text.append(element.get_text(strip=True))
    
    # 결과로 하나의 문자열을 반환
    return "\n".join(final_text)

# 새로운 컬럼 '질문내용_final'에 저장
df_q1['내용_body2_final'] = df_q1['내용_body2'].apply(extract_body_as_text_with_table)
```
