```python
# 감독관 컬럼을 타겟(y)으로 설정
y = df_input['감독관']

# 특성 데이터(x) 만들기
x = df_input[['감독관', '면적', '과수(감독전체)', '관할시군구수', '인구수', '사업체수', '종사자수', '신고사건접수건수', '신고사건처리건수', '평균처리일', '지연처리율', '근로감독건수', '정기감독실시업체',
              '수시감독실시업체', '사업장감독결과금품청산법위반건수', '현장점검횟수', '현장점검의날지도건수', '교육·컨설팅횟수', '안전보건감독건수', '산업재해조사표입력건수', '점검·감독' ,
              '중대재해', '사고재해자수', '사망만인율', '사고사망만인율', '부상자수', '재해자수', '재해율']]

```

```python
# 면적은 2차원 값이므로, 1차원 값인 거리를 추가
x['거리'] = x['면적'] ** 0.5
```

```python
excluded_columns_from_division = ['감독관', '지연처리율', '평균처리일', '사망만인율', '사고사망만인율', '재해율']
columns_to_divide = [col for col in x.columns if col not in excluded_columns_from_division]

# 3. 새로운 데이터프레임을 생성하여 '감독관'으로 나눈 값들을 저장
# .copy()를 사용하여 원본 x 데이터프레임이 수정되지 않도록 합니다.
x2 = x.copy()

# 4. '감독관' 컬럼의 값으로 각 행의 다른 모든 컬럼을 나눕니다.
# .div() 메서드를 사용하며, axis=0은 행 방향(즉, 각 행의 '감독관' 값으로)으로 나눔을 의미합니다.
# 주의: 만약 '감독관' 컬럼에 0이 있다면 ZeroDivisionError가 발생하거나 NaN(Not a Number)이 될 수 있습니다.
# 실제 데이터에서 이 부분을 어떻게 처리할지 고려해야 합니다 (예: 0을 다른 값으로 대체하거나 해당 행을 건너뛰기).
if '감독관' in x.columns:
    # '감독관' 컬럼이 0인 경우를 대비하여 조건부 나눗셈을 할 수도 있습니다.
    # 예시: x_normalized[columns_to_divide] = x[columns_to_divide].div(x['감독관'].replace(0, np.nan), axis=0)
    x2[columns_to_divide] = x[columns_to_divide].div(x['감독관'], axis=0)
else:
    print("오류: '감독관' 컬럼이 데이터프레임 x에 존재하지 않습니다.")
```

```python
x3= x2[['면적', '인구수', '사업체수', '종사자수', '신고사건접수건수', '신고사건처리건수', '평균처리일', '지연처리율', '근로감독건수', '정기감독실시업체',
        '수시감독실시업체', '사업장감독결과금품청산법위반건수', '현장점검횟수', '현장점검의날지도건수', '교육·컨설팅횟수', '안전보건감독건수', '산업재해조사표입력건수', '점검·감독' ,
        '중대재해', '사고재해자수', '사망만인율', '사고사망만인율', '부상자수', '재해자수', '재해율']]
```

```python
## LinearRegression 호출
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()

# .fit(독립 변수, 종속 변수) 메서드를 호출하여 모델을 학습시킵니다.
model.fit(x3, y)
```

```python
print("--- sklearn LinearRegression 결과 ---")

# 절편 (Intercept)
# model.intercept_ 에 절편값이 저장됩니다.
intercept = model.intercept_
print(f"✅ 절편 (b): {intercept:.4f}")

# 가중치 (Coefficients)
# model.coef_ 에 각 독립 변수의 가중치(계수)가 numpy 배열 형태로 저장됩니다.
coefficients = model.coef_
print("✅ 각 변수의 가중치 (w):")
# 가중치를 변수 이름과 매핑하여 출력합니다.
for feature_name, coef_value in zip(x3.columns, coefficients):
    print(f"   - {feature_name}: {coef_value:.4f}")

print("\n--- 수식 형태 ---")
# 추출된 값들을 사용하여 수식 형태를 출력
equation = f"y = {intercept:.4f}"
for i, feature_name in enumerate(x3.columns):
    equation += f" + {coefficients[i]:.4f} * {feature_name}"
print(equation)
```

```python
from sklearn.metrics import r2_score
import numpy as np

# --- 2. 전진 선택법(Forward Selection) 구현 ---

MAX_FEATURES = 7

# 최종 선택될 변수들을 저장할 리스트
selected_features = []
# 아직 모델에 포함되지 않은 (남아있는) 변수들 리스트
remaining_features = list(x3.columns)
# 모델 성능 평가 지표 (R-squared) 초기화
best_r2 = -float('inf') # 가능한 가장 낮은 값으로 초기화
# 더 이상 성능 개선이 없는지 판단하는 플래그
improved = True

print("--- 전진 선택법 시작 ---")

# 더 이상 개선이 없거나, 모든 변수를 다 확인했을 때까지 반복
while improved and remaining_features and len(selected_features) < MAX_FEATURES:
    improved = False # 이번 단계에서 성능 개선이 있었는지 추적
    temp_best_feature = None # 이번 단계에서 가장 좋았던 변수
    current_best_r2_step = best_r2 # 현재 단계의 최고 R-squared

    print(f"\n현재 선택된 변수: {selected_features}")
    print(f"남은 변수: {remaining_features}")

    # 남아있는 각 변수를 하나씩 추가해보면서 모델 성능 평가
    for feature_to_add in remaining_features:
        # 현재까지 선택된 변수들에 이번에 추가해볼 변수를 합침
        current_features_set = selected_features + [feature_to_add]

        # 선형 회귀 모델 학습
        model = LinearRegression()
        model.fit(x3[current_features_set], y)

        # 모델의 R-squared 값 계산
        y_pred = model.predict(x3[current_features_set])
        r2 = r2_score(y, y_pred)

        print(f"  '{feature_to_add}' 추가 시 R-squared: {r2:.4f}")

        # R-squared가 개선되면 최고 성능 및 변수 업데이트
        if r2 > current_best_r2_step:
            current_best_r2_step = r2
            temp_best_feature = feature_to_add
            improved = True # 성능 개선이 있었음을 표시

    # 이번 단계에서 성능이 개선되었다면, 해당 변수를 최종 선택 변수에 추가
    if improved:
        selected_features.append(temp_best_feature)
        remaining_features.remove(temp_best_feature) # 남은 변수에서 제거
        best_r2 = current_best_r2_step # 최고 R-squared 업데이트
        print(f"\n-> '{temp_best_feature}' 변수 선택됨. 새로운 최고 R-squared: {best_r2:.4f}")
    else:
        print("\n-> 더 이상 모델 성능을 개선하는 변수가 없습니다. 전진 선택법 종료.")

    if len(selected_features) == MAX_FEATURES:
        print(f"\n-> 최대 변수 개수 ({MAX_FEATURES}개)에 도달했습니다. 전진 선택법 종료.")
        break # 최대 변수 개수에 도달하면 루프 종료        

# --- 3. 최종 결과 확인 ---

print("\n--- 전진 선택법 최종 결과 ---")
print(f"최종 선택된 변수: {selected_features}")
print(f"최종 모델의 최고 R-squared: {best_r2:.4f}")

# 선택된 변수로 최종 모델 다시 학습하여 계수와 절편 확인
if selected_features:
    final_model = LinearRegression()
    final_model.fit(x3[selected_features], y)

    print("\n--- 최종 선택 변수를 사용한 모델의 회귀 계수 (가중치) 및 절편 ---")
    print(f"절편 (b): {final_model.intercept_:.4f}")
    print("각 변수의 가중치 (w):")
    for feature_name, coef_value in zip(selected_features, final_model.coef_):
        print(f"   - {feature_name}: {coef_value:.4f}")

    # 최종 모델의 수식 형태 출력
    equation = f"y = {final_model.intercept_:.4f}"
    for i, feature_name in enumerate(selected_features):
        equation += f" + {final_model.coef_[i]:.4f} * {feature_name}"
    print("\n최종 모델의 수식 형태:")
    print(equation)
else:
    print("전진 선택법으로 선택된 변수가 없습니다. 모델을 구성할 수 없습니다.")
```
