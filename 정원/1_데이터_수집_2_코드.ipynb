```python
file_path = '250523_code_table_v2.xlsx'

# 사용자가 언급한 시트 이름 목록입니다.
# 실제 엑셀 파일의 시트 이름과 정확히 일치해야 합니다.
target_sheet_names = [
    '공무원구분코드',
    '직군코드',
    '직렬코드',
    '직류코드',
    '계급코드',
    '직급코드',
    '직종코드'
]

# 각 DataFrame을 저장할 딕셔너리를 먼저 생성하는 것이 편리합니다.
dataframes = {}

try:
    # sheet_name=None으로 설정하면 모든 시트를 딕셔너리 형태로 한 번에 읽어옵니다.
    # 딕셔너리의 키는 시트 이름, 값은 해당 시트의 DataFrame이 됩니다.
    excel_file_content = pd.read_excel(file_path, sheet_name=None, dtype=str)

    print(f"'{file_path}' 파일에서 성공적으로 읽어온 시트 목록: {list(excel_file_content.keys())}")
    print("-" * 40)

    # 지정된 각 시트에 대해 DataFrame 변수를 생성하고 데이터를 할당합니다.
    # 변수 이름은 'df_' 접두사에 시트 이름을 붙여 만듭니다 (예: df_공무원구분코드).

    for sheet_name in target_sheet_names:
        if sheet_name in excel_file_content:
            # DataFrame 변수 생성 및 할당
            # globals()를 사용하여 동적으로 변수를 생성할 수도 있지만,
            # 여기서는 명시적으로 각 변수를 만들거나 딕셔너리에 저장하는 것을 보여드립니다.
            # 이 예제에서는 dataframes 딕셔너리에 저장합니다.
            dataframes[sheet_name] = excel_file_content[sheet_name]
            print(f"'{sheet_name}' 시트의 데이터를 DataFrame으로 읽어와 dataframes['{sheet_name}']에 저장했습니다.")
            # print(f"변수 df_{sheet_name} 에 저장했습니다.") # 만약 개별 변수를 원한다면 아래처럼 할당
            # globals()[f"df_{sheet_name}"] = excel_file_content[sheet_name]
        else:
            print(f"경고: '{sheet_name}' 시트를 파일에서 찾을 수 없습니다. 시트 이름을 확인해주세요.")
            dataframes[sheet_name] = pd.DataFrame() # 해당 시트가 없으면 빈 DataFrame으로 초기화

    # (선택 사항) 각 시트의 데이터를 별도의 변수로 직접 할당하고 싶다면:
    # 이 방법은 dataframes 딕셔너리를 사용하는 것보다 변수가 많아질 수 있습니다.
    df_공무원구분코드 = dataframes.get('공무원구분코드', pd.DataFrame())
    df_직군코드 = dataframes.get('직군코드', pd.DataFrame())
    df_직렬코드 = dataframes.get('직렬코드', pd.DataFrame())
    df_직류코드 = dataframes.get('직류코드', pd.DataFrame())
    df_계급코드 = dataframes.get('계급코드', pd.DataFrame())
    df_직급코드 = dataframes.get('직급코드', pd.DataFrame())
    df_직종코드 = dataframes.get('직종코드', pd.DataFrame())
    
except FileNotFoundError:
    print(f"오류: '{file_path}' 파일을 찾을 수 없습니다. 파일 경로를 정확하게 입력해주세요.")
except Exception as e:
    print(f"엑셀 파일을 읽는 중 오류가 발생했습니다: {e}")
    print("다음 사항을 확인해주세요:")
    print("1. 파일 경로가 올바른지 확인하세요.")
    print("2. 엑셀 파일이 손상되지 않았는지, 또는 올바른 엑셀 형식인지 확인하세요.")
    print("3. '.xlsx' 파일을 읽기 위해서는 'openpyxl' 라이브러리가 필요합니다. (설치: pip install openpyxl)")
```

```python
# 이전에 dataframes 딕셔셔너리가 생성되어 있고, 모든 컬럼이 텍스트로 로드되었다고 가정합니다.

# --- CSV 파일 저장 시작 ---
print("\n--- 중복 행 및 모든 값이 ''인 행 제거 후 CSV 파일 저장 시작 ---")

for sheet_name, df_original in dataframes.items():
    if not df_original.empty: # DataFrame이 비어있지 않은 경우에만 처리
        # 1. 중복된 행 제거
        df_after_duplicates = df_original.drop_duplicates(keep='first')
        rows_dropped_duplicates = df_original.shape[0] - df_after_duplicates.shape[0]

        if rows_dropped_duplicates > 0:
            print(f"'{sheet_name}' 시트 (중복 행 제거): {rows_dropped_duplicates}개 행 제거 완료. (중복 제거 후 행 수: {df_after_duplicates.shape[0]})")
        else:
            print(f"'{sheet_name}' 시트 (중복 행 제거): 중복된 행이 없습니다. (행 수: {df_after_duplicates.shape[0]})")

        # 2. 모든 컬럼의 값이 ''인 행 제거
        # (df == '')는 각 셀이 ''인지 여부를 boolean 값으로 표시하는 DataFrame을 만듭니다.
        # .all(axis=1)은 각 행에 대해 모든 값이 True인지(즉, 모든 셀이 '') 확인합니다.
        all_empty_string_rows_mask = df_after_duplicates.eq('').all(axis=1)
        df_final_cleaned = df_after_duplicates[~all_empty_string_rows_mask]

        rows_all_empty = df_after_duplicates.shape[0] - df_final_cleaned.shape[0]

        if rows_all_empty > 0:
            print(f"'{sheet_name}' 시트 (모든 값 '' 행 제거): {rows_all_empty}개 행 제거 완료. (최종 저장될 행 수: {df_final_cleaned.shape[0]})")
        else:
            print(f"'{sheet_name}' 시트 (모든 값 '' 행 제거): 모든 값이 ''인 행이 없습니다. (최종 저장될 행 수: {df_final_cleaned.shape[0]})")


        csv_file_name = f"{sheet_name}.csv"
        try:
            # 최종 정제된 DataFrame을 CSV 파일로 저장
            df_final_cleaned.to_csv(csv_file_name, index=False, encoding='utf-8-sig')
            print(f"'{csv_file_name}' 파일로 성공적으로 저장되었습니다.")
        except Exception as e:
            print(f"'{csv_file_name}' 파일 저장 중 오류 발생: {e}")
        print("-" * 30) # 각 파일 처리 후 구분선
    else:
        print(f"'{sheet_name}'에 해당하는 DataFrame이 비어있어 처리 및 CSV 저장을 건너뜁니다.")
        print("-" * 30)


print("-" * 40)
print("모든 DataFrame에 대해 정제 작업 시도 및 CSV 파일 저장 작업을 완료했습니다.")
```

```python
# 이전에 df_공무원구분코드, df_직군코드 ... 등의 변수 또는
# dataframes 딕셔셔너리가 생성되어 있고, 모든 컬럼이 텍스트로 로드되었다고 가정합니다.

# --- CSV 파일 저장 시작 ---
print("\n--- 중복 행 제거 후 CSV 파일 저장 시작 ---")

# 방법 1: dataframes 딕셔너리를 순회하며 처리 및 저장
for sheet_name, df in dataframes.items():
    if not df.empty: # DataFrame이 비어있지 않은 경우에만 처리
        # 중복된 행 제거 (기본적으로 모든 컬럼을 비교하여 중복 여부 판단, 첫 번째 행 유지)
        df_cleaned_rows = df.drop_duplicates(keep='first')

        # 중복 제거 전후 행 수 비교 (선택 사항)
        if df.shape[0] != df_cleaned_rows.shape[0]:
            rows_dropped = df.shape[0] - df_cleaned_rows.shape[0]
            print(f"'{sheet_name}' 시트: 중복 행 {rows_dropped}개 제거 완료. (원본 행: {df.shape[0]}, 저장될 행: {df_cleaned_rows.shape[0]})")
        else:
            print(f"'{sheet_name}' 시트: 중복된 행이 없습니다. (행 수: {df_cleaned_rows.shape[0]})")

        csv_file_name = f"{sheet_name}.csv"
        try:
            # 중복 행이 제거된 DataFrame을 CSV 파일로 저장
            # index=False: DataFrame 인덱스를 CSV 파일에 포함하지 않음
            # encoding='utf-8-sig': Excel에서 한글 깨짐 방지
            df_cleaned_rows.to_csv(csv_file_name, index=False, encoding='utf-8-sig')
            print(f"'{csv_file_name}' 파일로 성공적으로 저장되었습니다.")
        except Exception as e:
            print(f"'{csv_file_name}' 파일 저장 중 오류 발생: {e}")
        print("-" * 20) # 각 파일 처리 후 구분선
    else:
        print(f"'{sheet_name}'에 해당하는 DataFrame이 비어있어 처리 및 CSV 저장을 건너뜁니다.")

# 방법 2: 개별 변수를 직접 지정하여 처리 및 저장 (만약 개별 변수를 주로 사용하신다면)
# print("\n--- 중복 행 제거 후 CSV 파일 저장 시작 (개별 변수 사용) ---")
#
# # 예시: df_공무원구분코드 처리
# if 'df_공무원구분코드' in locals() and not df_공무원구분코드.empty:
#     original_row_count = df_공무원구분코드.shape[0]
#     df_cleaned = df_공무원구분코드.drop_duplicates(keep='first')
#     cleaned_row_count = df_cleaned.shape[0]
#
#     if original_row_count != cleaned_row_count:
#         print(f"'df_공무원구분코드': 중복 행 {original_row_count - cleaned_row_count}개 제거. (원본: {original_row_count}, 저장: {cleaned_row_count})")
#     else:
#         print(f"'df_공무원구분코드': 중복된 행이 없습니다. (행 수: {cleaned_row_count})")
#     try:
#         df_cleaned.to_csv('공무원구분코드.csv', index=False, encoding='utf-8-sig')
#         print("'공무원구분코드.csv' 파일로 성공적으로 저장되었습니다.")
#     except Exception as e:
#         print(f"'공무원구분코드.csv' 파일 저장 중 오류 발생: {e}")
# else:
#     print("'df_공무원구분코드'가 비어있거나 정의되지 않아 처리 및 CSV 저장을 건너뜁니다.")
#
# # ... (df_직군코드, df_직렬코드 등 나머지 개별 변수들에 대해서도 동일하게 반복) ...

print("-" * 40)
print("모든 DataFrame에 대해 중복 행 제거 시도 및 CSV 파일 저장 작업을 완료했습니다.")
```
