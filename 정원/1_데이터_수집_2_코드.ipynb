```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None) 
```

```python
# 파일 경로 지정
file_path = 'd:/Workspace/data/ldap_code_full.txt'

try:
    # 지정된 경로의 파일을 읽는다.
    ldap_code = pd.read_csv(file_path, sep='\t', encoding='cp949')
except FileNotFoundError:
    print(f"오류: '{file_path}' 파일을 찾을 수 없습니다. 파일 경로를 확인해주세요.")
except pd.errors.EmptyDataError:
    print(f"오류: '{file_path}' 파일이 비어 있습니다.")
except Exception as e:
    print(f"파일을 읽는 중 오류가 발생했습니다: {e}")
```

```python
# 읽은 내용 확인
ldap_code.head()
```

```python
# 컬럼명 확인
ldap_code.columns
```

```python
# '전체기관명' 컬럼에 '고용노동부'가 포함된 행만 선택한다.
filtered_ldap_code_contains = ldap_code[ldap_code['전체기관명'].str.contains('중앙노동위원회', na=False)]

# 필터링된 결과를 출력합니다.
if not filtered_ldap_code_contains.empty:
    print("\n'전체기관명'이 '고용노동부'인 데이터:")
    print(filtered_ldap_code_contains)
else:
    print("\n'전체기관명'이 '고용노동부'인 데이터를 찾을 수 없습니다.")
```
