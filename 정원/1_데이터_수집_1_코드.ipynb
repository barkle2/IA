```python
import pandas as pd
import os
```

```python
# 파일 경로 지정
file_path = 'd:/Workspace/data/ldap_code_full.txt'

try:
    # 지정된 경로의 파일을 읽는다.
    ldap_code = pd.read_csv(file_path, sep='\t', encoding='cp949')
except FileNotFoundError:
    print(f"오류: '{file_path}' 파일을 찾을 수 없습니다. 파일 경로를 확인해주세요.")
except pd.errors.EmptyDataError:
    print(f"오류: '{file_path}' 파일이 비어 있습니다.")
except Exception as e:
    print(f"파일을 읽는 중 오류가 발생했습니다: {e}")
```

```python
# CSV 파일을 저장할 디렉토리 (선택 사항, 없으면 스크립트와 같은 위치에 저장)
output_directory = "code_table"

# 출력 디렉토리가 없으면 생성
if not os.path.exists(output_directory):
    os.makedirs(output_directory)
    print(f"'{output_directory}' 디렉토리를 생성했습니다.")

# 출력할 CSV 파일 경로 (시트 이름을 파일명으로 사용)
# os.path.join을 사용하여 운영체제에 맞는 경로 구분자로 파일 경로를 만듭니다.
csv_file_name = 'el_ldap_code.csv'
csv_file_path = os.path.join(output_directory, csv_file_name)

# 데이터프레임을 CSV 파일로 출력 (인덱스 제외)
el_ldap_code.to_csv(csv_file_path, index=False, encoding='utf-8-sig')
print(f"'{sheet_name}' 시트의 데이터프레임을 '{csv_file_path}' 파일로 저장했습니다.")
```

```python
# 엑셀 파일 경로
excel_file_path = "250523_code_table_v2.xlsx"
# 시트 이름 리스트
sheet_names = ['공무원구분코드', '직군코드', '직렬코드', '직류코드', '계급코드', '직급코드', '직종코드']

# 엑셀 파일 자체를 한번만 열기 (성능 향상 가능)
xls = pd.ExcelFile(excel_file_path)

# 각 시트 이름을 순회하며 작업 수행
for sheet_name in sheet_names:
    print(f"\n--- '{sheet_name}' 시트 처리 중 ---")
    try:
        # 지정된 시트 이름을 사용하여 데이터프레임으로 읽어오기
        dataframe = pd.read_excel(xls, sheet_name=sheet_name, dtype=str)        

        # 중복된 행 확인
        duplicate_rows = dataframe.duplicated()
        num_duplicate_rows = duplicate_rows.sum()

        if num_duplicate_rows > 0:
            print(f"'{sheet_name}' 시트에 총 {num_duplicate_rows}개의 중복된 행이 있습니다.")
            # 중복된 행 제거 (첫 번째 행 유지)
            dataframe_without_duplicates = dataframe.drop_duplicates()
            print("중복된 행을 제거했습니다.")
        else:
            print(f"'{sheet_name}' 시트에 중복된 행이 없습니다.")
            dataframe_without_duplicates = dataframe

        # 출력할 CSV 파일 경로 (시트 이름을 파일명으로 사용)
        # os.path.join을 사용하여 운영체제에 맞는 경로 구분자로 파일 경로를 만듭니다.
        csv_file_name = f"{sheet_name}.csv"
        csv_file_path = os.path.join(output_directory, csv_file_name)

        # 데이터프레임을 CSV 파일로 출력 (인덱스 제외)
        dataframe_without_duplicates.to_csv(csv_file_path, index=False, encoding='utf-8-sig')
        print(f"'{sheet_name}' 시트의 데이터프레임을 '{csv_file_path}' 파일로 저장했습니다.")

    except KeyError:
        print(f"오류: '{excel_file_path}' 파일에 '{sheet_name}' 시트를 찾을 수 없습니다. 시트 이름을 확인해주세요.")
    except Exception as e:
        print(f"'{sheet_name}' 시트 처리 중 오류가 발생했습니다: {e}")
```
